```python
import pandas as pd
kdf = pd.read_pickle('use_embeddings.p')
gdf = pd.read_csv('gdf.csv')
```

```python
def get_use_cases(company, description, users, data):
  conversational_use_cases = llm.predict(
        f"""
        INSTRUCTIONS:
        You are an expert consultant in the healthcare and technology industry. You have years of experience working in business operations and engineering at major health tech companies.
        Given a company's name, description, target users, and data, make an educated guess on what kinds of use cases this company may have for a conversational chatbot, either internally or externally.
        Then, format into the xml format provided below.

        FORMAT:
        <output>
          <company>company name</company>
          <use_cases>
            <use_case id=0>
              <name>name of the use case. It should be VERY SPECIFIC.</name>
              <description>a brief one-sentence description of the use case</description>
              <data>data that the use case would work with (can be multiple)</data>
              <users>intended users of the chatbot (can be multiple)</users>
            </use_case>
            ... multiple cases
          </use_cases>
        </output>

        INPUT:
        <input>
          <company>{company}</company>
          <description>{description}</description>
          <users>{users}</users>
          <data>{data}</data>
        </input>

        Remember, it MUST be WELL-FORMED XML.

        OUTPUT: """)
  return conversational_use_cases
```

```python
cases = []

for row in gdf.iterrows():
    name = row[1]['name']
    description = row[1]['description']
    users = row[1]['users']
    data = row[1]['data']
    conversational_use_cases = get_use_cases(name, description, users, data)

    cases += [conversational_use_cases]
```

```python
import json
open('./cases.txt', 'w').write(json.dumps(cases))
```

```python
import xmltodict
import re

ds = []
failed = []

for case in cases:
    try:
        case = re.sub(r'^\s*', '', case, flags=re.M)
        case = re.sub(r'\n', '', case)
        d = xmltodict.parse(case)
        ds += [d]
    except:
        failed += [case]
        continue
```
